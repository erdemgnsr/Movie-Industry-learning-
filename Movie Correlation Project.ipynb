```python

#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_rows",200)
```

```python
#read data
data = pd.read_csv("movies.csv")
```

```python
#movies has biggest gross
data.sort_values("gross",ascending = False)
```

```python
data.shape==data.drop_duplicates().shape
```

```python
data["name"].shape == data["name"].drop_duplicates().shape
```

```python
#higher company income
data.groupby(by="company")["gross"].sum().sort_values(ascending=False)
```

```python
#Budget vs Gross scatters
plt.figure(figsize = (15,10))
plt.scatter(data["budget"],data["gross"])
plt.xlabel("Budget")
plt.ylabel("Gross")
plt.title("Budget vs Gross")
plt.show()
```

```python
#budget  vs gross with regression

sns.regplot(x = "budget", y = "gross" ,data = data, scatter_kws = {"color" : "r"}, line_kws = {"color" : "b"}).figure.set_size_inches(15,10)
```

```python
#correlations between not object datatypes      #pearson kendall spearman callable

data.corr() #pearson
```

```python
data.corr(method = "spearman")
```

```python
data.corr(method = "kendall")
```

```python
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix,annot = True).figure.set_size_inches(13,10)
plt.show()
```

```python
numerized_data = data

for column_name in numerized_data.columns:
    if (numerized_data[column_name].dtype == "object"):
        numerized_data[column_name] = numerized_data[column_name].astype("category")
        numerized_data[column_name] = numerized_data[column_name].cat.codes
numerized_data.head(200)
```

```python
numerized = data

for column_name in numerized_data.columns:
    if (numerized[column_name].dtype == "object"):
        numerized[column_name] = numerized[column_name].astype("category")

numerized.head(200)
```
